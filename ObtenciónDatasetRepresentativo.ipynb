```python
import dask.dataframe as dd
import pandas as pd

file_path = '/content/drive/MyDrive/used_cars_data.csv'
stratify_column = 'make_name'
sample_size = 150000
output_file = 'muestra_estratificada.csv'

dtype_fix = {'bed': 'object',
             'bed_height': 'object',
             'bed_length': 'object',
             'cabin': 'object',
             'sp_id': 'float64',
             'dealer_zip': 'object'}

# 1. Leer el archivo grande con Dask
ddf = dd.read_csv(file_path, dtype=dtype_fix)

# 2. Calcular la fracción
total_rows = len(ddf)
fraction = sample_size / total_rows

# 3. Realizar el muestreo estratificado
# Le decimos a dask que agrupe y luego APLIQUE una función de muestreo a cada grupo.
# El parámetro 'meta' le dice a Dask cómo se verá el resultado sin tener que calcularlo todo.
sample_ddf = ddf.groupby(stratify_column).apply(
    lambda x: x.sample(frac=fraction, random_state=42),
    meta=ddf._meta
)

# 4. Guardar el resultado
sample_ddf.to_csv(output_file, single_file=True, index=False)

print(f"¡Listo! Muestra estratificada por '{stratify_column}' ha sido guardada en '{output_file}'")
```
